```python
import pandas as pd
```

```python
corpus = pd.DataFrame(pd.read_table('./urdu-sentiment-corpus-v1.tsv'))
corpus.head()
```

```python
corpus = corpus[corpus['Class'] != 'O']
corpus.head()
```

```python
from sklearn.model_selection import train_test_split

train_data , test_data = train_test_split(corpus, test_size=0.25, random_state=42)

print(train_data.shape,"___",test_data.shape)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import SimpleRNN, GRU, LSTM, Bidirectional, Dense

num_words = 3000
oov_token = '<UNK>'
pad_type = 'post'
trunc_type = 'post'


tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(train_data.Tweet)

# Get our training data word index
word_index = tokenizer.word_index

# Encode training data sentences into sequences
train_sequences = tokenizer.texts_to_sequences(train_data.Tweet)

# Get max training sequence length
train_maxlen = max([len(x) for x in train_sequences])

# Pad the training sequences
x_train  = pad_sequences(train_sequences, padding=pad_type, truncating=trunc_type, maxlen=30)


# Output the results of our work
print("Word index:\n", word_index)
print("\nTraining sequences:\n", train_sequences)
print("\nPadded training sequences:\n", x_train )
print("\nPadded training shape:", x_train .shape)
print("Training sequences data type:", type(train_sequences))
print("Padded Training sequences data type:", type(x_train ))
```

```python
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(test_data.Tweet)

# Get our testing data word index
word_index = tokenizer.word_index

# Encode test data sentences into sequences
test_sequences = tokenizer.texts_to_sequences(test_data.Tweet)

# Get max testing sequence length
test_maxlen = max([len(x) for x in test_sequences])

# Pad the testing sequences
X_test = pad_sequences(test_sequences , padding=pad_type, truncating=trunc_type, maxlen=30)

# Output the results of our work
print("Word index:\n", word_index)
print("\nTesting sequences:\n", test_sequences)
print("\nPadded testing sequences:\n", X_test )
print("\nPadded testing shape:", X_test .shape)
print("Testing sequences data type:", type(test_sequences))
print("Padded Testing sequences data type:", type(X_test))
```

```python
for x, y in zip(train_data.Tweet, x_train):
  print('{} -> {}'.format(x, y))

print("\nWord index (for reference):", word_index)
```

```python
labels = ['P', 'N']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(train_data.Class.to_list())
y_train = encoder.transform(train_data.Class.to_list())
y_test = encoder.transform(test_data.Class.to_list())
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)

print(y_train.shape, "____", y_test.shape)
print(x_train, "____",y_train)
```

```python
from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python

# Define the model architecture
model = Sequential()
model.add(SimpleRNN(30, input_shape=(30,1) ,return_sequences=True, dropout=0.3, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(SimpleRNN(15,dropout=0.3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy',recall_m,precision_m,f1_m])

history = model.fit(x_train, y_train, epochs=10, batch_size=64)

loss, accuracy, recall, precision, f1_score = model.evaluate(X_test, y_test, verbose=0)
print('Loss:', loss)
print('Accuracy:', accuracy)
print('f1_score:', f1_score)
print('precision:', precision)
print('recall:', recall)
```
